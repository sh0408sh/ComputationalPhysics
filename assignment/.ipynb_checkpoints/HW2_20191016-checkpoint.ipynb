Question 1. Without using built-in functions (e.g. sort() or sorted()), make
your own script which gets a string and rearrange the letters of the string in
an alphabetical order. Make it case-insensitive.

```python
# inpyt the string
s = input('string :')

# type casting to list to deal with easy 
s2 = list(s.lower())
s = list(s)

"""
Using Selecting sort algorithm : find minimum for each loop and set minimum first. 
since we consider case-insentive case, I will compare the alphabet order in s2 case(all characters are lower ; 'A' and 'a' have different
ascii code. so we should covert all characters to lower for case insentive comparison.) 
Both s2 and s have swapping. s2 should be swapped for sorting and s should be swapped since it is real string we are dealing with. 
"""

for i in range(len(s)) :
    for j in range(i+1,len(s)) :
        if s2[i] > s2[j] :
            tmp = s[i] ; s[i] = s[j] ; s[j] =  tmp
            tmp2 = s2[i] ; s2[i] =  s2[j] ; s2[j] = tmp2

#convert list to string
s_final = ''
            
for i in range(len(s)) :
    s_final += s[i]
            
print(s_final)
```

### Newton Method 
[Newton's method](https://patrickwalls.github.io/mathematicalpython/root-finding/newton/)

root finding method using linear approximation. 
1. Choosing one point x0
2. Draw a tangent for that point y = f'(x0)(x-x0) + f(x0)
3. Find x-intersection of tangent f'(x0)(x1-x0) + f(x0) = 0 -> x1 = x0 -f(x0)/f'(x0)
4. Repeat it until the error go small we want.

```python
# define function to find solution using Newton method.
# f : function, x0 : starting point  err : Stopping criteria is abs(f(x)) < epsilon. n : the number of solution we find
def newton(f, x0,err, n ) :
    xn = x0
    solution = []
    xns =[]
    fxns =[]
    for i in range(n) :
        # loop for 4
        while abs(f(xn)) > err :
            xns.append(xn)
            fxns.append(f(xn))
            tmp = xn
            # process 3 : find next point(x-intersection)
            xn = xn -f(xn)/sm.derivative(f,xn, dx = 1e-6)
        solution.append(xn)
        # if we find solution, it will satisfy error condition, so xn will not change. 
        # Therefore, we will change xn appropriately to find next consecutive root. 
        """
        Actually I don't know what is specific(mathematical) criteria to move xn for consecutive roots. 
        If moving interval is too small, then it comes back to former value we already found.
        Some moving interval makes next xn too far which is not consecutive.
        Therefore, I drew graph, estimated solution value, and adjusted interval.
        """
        xn = xn + 0.5
        if i>=2 :
            xn = xn -0.15
    #print(xns)
    #print(fxns)
        
    
    return solution
    
```

Question 3. Find the solution of an equation erf(ùë•) = 0.5 by using any root
finder. Here erf(ùë•) is an error function defined by $erf(ùë•) = {{2}\over{\sqrt{\pi}}}\int^{x}_{0}e^{-t^2}$
ùëëùë°. Do
not use a built-in error function, but create your own using trapezoid method.

```python
#define erf(t)

def erf(t) :
    x = np.linspace(0,t,1000)
    y = np.exp(-(x**2))
    # Represent integration as trapzoid method 
    erfx = (2/np.sqrt(np.pi))*np.trapz(y,x)
    return erfx
    
```

```python
"""
question to chat GPT : Find the code for solution of an equation erf(ùë•) = 0.5 by using any root
finder. Here erf(ùë•) is an error function defined by $erf(ùë•) = {{2}\over{\sqrt{\pi}}}\int^{x}_{0}e^{-t^2}$
ùëëùë°. Do not use a built-in error function, but create your own using trapezoid method.
"""
```

```python
import math

# Define integration using trapezoid method 
def trapezoid(f, a, b, n):
    h = (b - a) / n
    x = [a + i * h for i in range(n+1)]
    y = [f(xi) for xi in x]
    return h/2 * (y[0] + y[n] + 2 * sum(y[1:n]))

# Define error function
def error_function(x):
    f = lambda t: math.exp(-t**2)
    return 2 / math.sqrt(math.pi) * trapezoid(f, 0, x, 1000)
```

```python
# find the solution of erf(x) = 0.5
from scipy.optimize import bisect

x = bisect(lambda x: error_function(x) - 0.5, 0, 2)
print("x = ", x)
```
