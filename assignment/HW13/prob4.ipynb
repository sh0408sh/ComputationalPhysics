```python
import numpy as np
import scipy.optimize as opt
```

```python
import matplotlib.pyplot as plt
import scipy.misc as sm
```

```python
def f(x):
	y=2*x*x-7*x+1
	return y
```

```python
x=np.arange(0,10,0.1)
y = 2*x*x-7*x+1
```

```python
def Newton(f,x0,err) :
    xn = x0
    while abs(f(xn)) > err :
        tmp = xn
        xn = xn - f(xn)/sm.derivative(f,xn,dx = 1e-6)
    return xn
```

```python
sol1 = Newton(f,1,1e-3)
sol1
```

```python
sol2 = Newton(f,10,1e-3)
sol2
```

```python
f(sol2)
```
