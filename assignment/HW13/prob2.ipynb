```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(0,9,num=10)
y=np.array([1.1,2.2,3.1,2.9,4.1,6.7,9.8,11.1,13.2,15.3])
```

### Least square method:

Let us consider a general linear functional $L$:
\begin{equation}
    L({\bf a,x})=x_0+a_1x_1+a_2x_2+\cdots+a_px_p
\end{equation}

For given $({\bf a_i},y_i),i=1,\cdots,n$, find $ {\bf x}$ to minimize the following
$$\sum_{i=1}^n||L({\bf a_i,x_i})-{ y_i}||^2_2$$


* $L({\bf a_i,x_i})\approx { y_i} \Rightarrow A X \approx Y \Rightarrow \hat{X}=(A^T A)^{-1}A^T Y$ 

```python
# Write your own linear fitter here

# assemble matrix A
A = np.vstack([x, np.ones(len(x))]).T

# turn y into a column vector
y = y[:, np.newaxis]
```

```python
# Direct least square regression
alpha = np.dot((np.dot(np.linalg.inv(np.dot(A.T,A)),A.T)),y)
print(alpha)
```

```python
# Make a plot of your fitting line here
plt.plot(x,y,'o')
plt.plot(x, alpha[0]*x + alpha[1], 'r')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
