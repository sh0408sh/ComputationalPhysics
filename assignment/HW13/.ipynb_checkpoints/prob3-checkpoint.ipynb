```python
import random
import scipy.signal as sg
import numpy as np
import matplotlib.pyplot as plt
```

```python
# this is a signal with random noise
x=np.arange(100)
y0=np.sin(10.0*np.pi*x/100)
y=np.sin(10.0*np.pi*x/100)
```

```python
for i in x:
	y[i] += 2.0*(random.random()-0.5)
```

```python
plt.plot(x,y); plt.show()
##############################3
```

```python
# Write your own sav-gol filter here and plot it along with the original signal
y2 = y*0
y2[0] =  0.486*y[0] +0.343*y[1]-0.086*y[2]
y2[1] = 0.343*y[0]+0.486*y[1] +0.343*y[2]-0.086*y[3]
y2[98] = -0.086*y[96] +0.343*y[97]+0.486*y[98] +0.343*y[99]
y2[99] = -0.086*y[97] +0.343*y[98]+0.486*y[99]
for i in range(2,100-2) :
        y2[i] = -0.086*y[i-2] +0.343*y[i-1]+0.486*y[i] +0.343*y[i+1]-0.086*y[i+2]
```

```python
plt.plot(x,y)
plt.plot(x,y2); plt.show()
```
