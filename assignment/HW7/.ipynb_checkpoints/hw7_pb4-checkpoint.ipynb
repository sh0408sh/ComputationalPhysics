Question 4. Find up to three energy levels from the 
ground state of a particle trapped in a onedimensional, asymmetric potential well as in the 
figure. 



```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integ
```

```python
h=0.005
L = 0.5
t=np.arange(0,L*2,h)
```

```python
#right side
def fr(y,t,E):
	x,v=y
	if t<L:
		U=0.0
	else:
		U=5.0
	return [v,2.0*(U-E)*x]
```

```python
#left side
def fl(y,t,E):
	x,v=y
	if t<L:
		U=0.0
	else:
		U=10.0
	return [v,2.0*(U-E)*x]
```

```python
#initial condition for assymetricity
y0=[0.0,1.0]
```

```python
E = 0
h = 1e-2
while True : 
    solr = integ.odeint(fr,y0,t,args=(E,))
    soll = integ.odeint(fl,y0,t,args=(E,))
    soll = soll[::-1]
    if sol[-1,0]<0:
            plt.plot(t,sol[:,0]); plt.show()
            break
    E = E + h 
```

```python
E = 2
sol = integ.odeint(fr,y0,t,args=(E,))
plt.plot(t,sol[:,0]); plt.show()
```

```python
E =8.7
sol = integ.odeint(fl,y0,t,args=(E,))
plt.plot(t,sol[:,0]); plt.show()
```
