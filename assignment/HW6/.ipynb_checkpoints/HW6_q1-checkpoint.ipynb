Question 1. Using ‘ode-Kepler.py’, verify the Kepler’s 3rd law for five different 
elliptical orbits.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as integ
```

```python
# A vector function. y is a two-compo vector, i.e. (x,v).
# f returns two-compo vector, i.e. (f1,f2)=(.
def F(t,z):  
	x,y,vx,vy=z  
	r=np.sqrt(x**2+y**2)
	fx=-x/r**3; fy=-y/r**3
	return [vx,vy,fx,fy] 
```

```python
z0=[1.0, 0.0, 0.00, 0.5] #x0,y0,vx0,vy0
```

```python
def orbit( z0) :
    h=float(0.001)   # time step
    nStep=int(10000)  # num. steps to run
    t=np.arange(0,h*nStep,h)  # an array of discretized time 
    #sol = integ.odeint(F,z0,t) # sol is [[x0,y0,vx0,vy0],..]
    sol = integ.solve_ivp(F,[0,10],z0,method='RK45',t_eval=t) # sol is [[x0,y0,vx0,vy0],..]
    
    
#     index = np.where(sol.y[0]>=0 and sol.y[2]>=0)

    
    for i in range(1, len(sol.y[0])-1) : # for x coordinate 
        if sol.y[0][i]>=0 and sol.y[2][i]>=0 : # when x > =0  and vx >=0
            if sol.y[1][i]*sol.y[1][i+1] <= 0 : # when y sign changes
                index = i
                t = sol.t[i] #period
                break

    plt.plot(sol.y[0],sol.y[1]); plt.show()
    
    print(sol.y[0][index],sol.y[1][index],sol.y[2][index],sol.y[3][index]) #coordinates for t
    
    return t**2/z0[0]**3 #check the Kepler's 3rd law
    
```

```python
orbit([1.0, 0.0, 0.00, 0.5])
```

```python
orbit([1.001, 0.0, 0.00, 0.5])
```

```python
orbit([1.002, 0.0, 0.00, 0.5])
```

```python
orbit([1.003, 0.0, 0.00, 0.5])
```
