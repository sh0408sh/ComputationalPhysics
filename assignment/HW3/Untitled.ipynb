```python
xmax=float(2)
dx=float(0.01)
dt=float(0.005)
f=float(2)
smax=int(1257)
n1 = 1
n2 = 1.3
```

```python
a=dt/dx
w=2.0*np.pi*f
```

```python
x=np.arange(0,xmax+dx,dx)
c=int(0.5*xmax/dx)
```

```python
Eiy=0*x; Eiz=0*x
Biy=0*x; Biz=0*x
```

```python
Ety=0*x; Etz=0*x
Bty=0*x; Btz=0*x
```

```python
Ery=0*x; Erz=0*x
Bry=0*x; Brz=0*x
```

```python
Ey[c:]
```

```python
s=0
while s<smax:
	Biy[:-1]  +=  a*(Eiz[1:]-Eiz[:-1])
	Biz[:-1]  += -a*(Eiy[1:]-Eiy[:-1]) 
	Eiy[1:-1] += -a*(Biz[1:-1]-Biz[0:-2]) 
	Eiy[0] += -dt*np.sin(w*s*dt)
	Eiz[1:-1] +=  a*(Biy[1:-1]-Biy[0:-2])
	s+=1
```

```python
s=0
while s<smax:
	By[:-1]  +=  a*(Ez[1:]-Ez[:-1])
	Bz[:-1]  += -a*(Ey[1:]-Ey[:-1]) 
	Ey[1:-1] += -a*(Bz[1:-1]-Bz[0:-2]) 
	Ey[c] += -dt*np.sin(w*s*dt)
	Ez[1:-1] +=  a*(By[1:-1]-By[0:-2])
	s+=1
```
